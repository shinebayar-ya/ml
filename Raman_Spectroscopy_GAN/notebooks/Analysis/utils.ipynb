```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../input/data.csv")
```

```python
X = data.drop(['Diagnostic', 'COVID-19_Status'], axis=1)
X.columns = X.columns.astype(int)
y_diagnostic = data['Diagnostic']
y_covid_status = data['COVID-19_Status']
```

```python
wavelength = X.columns
intensity = X.values
```

```python
fig, ax = plt.subplots(figsize=(18, 6))

sns.lineplot(x=wavelength, y=intensity[0], ax=ax)
ax.set_xlabel(' ')
ax.set_ylabel('Intensity')
ax.legend()

plt.savefig('../figures/spectrum.png', dpi=600, bbox_inches='tight')
plt.show()
```
